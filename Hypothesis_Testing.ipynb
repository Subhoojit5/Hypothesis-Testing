# Question 1

Null Hypothesis: There is no significant difference in the diameter of the cutlet between two units.

Alternate Hypothesis:There is a significant difference in the diameter of the cutlet between two units.

```python
import pandas as pd
from scipy import stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
Cutlets=pd.read_csv('Cutlets.csv')
Cutlets.head()
```

```python
results=stats.ttest_ind(Cutlets['Unit A'],Cutlets['Unit B'])
results
```

```python
p_value=results[1]
p_value
```

Since the P-value is greater than the significance level(0.05), we can go with the null hypothesis (No significant difference in the diameter of the cutlet between two units)

# Question 2

Null Hypothesis: Average is same for all the 4 laboratories

Alternate Hypothesis: Average is not same. Ateast one laboratory average is different.

In this case we will use ANOVA

```python
import pandas as pd
from scipy import stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('LabTAT.csv')
data
```

```python
results=stats.f_oneway(data['Laboratory 1'],data['Laboratory 2'],data['Laboratory 3'],data['Laboratory 4'])
results
```

```python
p_value=results[1]
p_value
```

Since the P-value(2.11e-57) is very much less than the significance level(0.05), we can go with the **alternate hypothesis** (Average is not same. Ateast one laboratory average is different).

# Question 3

Null Hypothesis: The male-female buyer ratio is similar across all region.

Alternate Hypothesis: The male-female buyer ratio is not similar across all region.

In this case we will use Chi-Square, to see whether the male-female variables are dependent or independent.

```python
import pandas as pd
from scipy import stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('BuyerRatio.csv')
data
```

```python
df=data.iloc[:,1:5]
df
```

```python
results=stats.chi2_contingency(df)
results
```

```python
p_value=results[1]
p_value
```

Null Hypothesis will be accepted since the p-value is greater than the significant value(0.05)

# Question 4

Alternate Hypothesis: The defective % does varies by the centre.

```python
import pandas as pd
from scipy import stats
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('Costomer+OrderForm.csv')
data
```

```python
results=stats.chi2_contingency([data['Phillippines'].value_counts(),data['Indonesia'].value_counts(),data['Malta'].value_counts(),data['India'].value_counts()])
results
```

```python
p_value=results[1]
p_value
```
